```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Defining the periodic function
def f(x):
    return np.where(x < 0, x**2 - 1, x * np.exp(-x/2))

# Generating x values for the plot
x = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Evaluating the function for each x value
y = [f(xi) for xi in x]

# Creating a figure for the plot
plt.figure(figsize=(8, 6))

# Plotting the periodic function
plt.plot(x, y)

# Adding title and labels to the plot
plt.title("Periodic Function f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")

# Displaying the grid on the plot
plt.grid(True)

# Showing the plot
plt.show()
```

# Q1 b & c

```python
# Importing necessary libraries
from scipy import integrate

# Function for the periodic function to be approximated
def f(x):
    return np.where(x < 0, x**2 - 1, x * np.exp(-x/2))

# Function to compute the Fourier series approximation
def fourier_series(x, n):
    # Computing the constant term (a0) of the Fourier series
    a0 = integrate.quad(f, -np.pi, np.pi)[0] / (2 * np.pi)
    
    # Initializing an array to accumulate Fourier series terms
    result = np.zeros_like(x)
    
    # Looping over harmonics to calculate coefficients and accumulate terms
    for i in range(1, n+1):
        # Calculating coefficients for cosine and sine terms
        a_n = (1/np.pi) * integrate.quad(lambda x: f(x) * np.cos(i*x), -np.pi, np.pi, limit=1000)[0]
        b_n = (1/np.pi) * integrate.quad(lambda x: f(x) * np.sin(i*x), -np.pi, np.pi, limit=1000)[0]
        
        # Accumulating Fourier series terms
        result += a_n * np.cos(i*x) + b_n * np.sin(i*x)
    
    # Returning the computed Fourier series approximation
    return a0/2 + result

# Plotting Fourier series expansions for different harmonics
for n in [1, 5, 150]:
    plt.figure(figsize=(10, 6))
    plt.plot(x, fourier_series(x, n))
    plt.title(f'Fourier series expansion up to {n}th harmonic')
    plt.show()
```

# Q1 d

```python
# Define the RMSE function
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred)**2))

# Calculate and plot the RMSE for each harmonic
x = np.linspace(-4 * np.pi, 4 * np.pi, 1000)
y_true = f(x)
rmses = []

# Loop through harmonics to compute RMSE for each
for n in range(151):
    y_pred = fourier_series(x, n)
    error = rmse(y_true, y_pred)
    rmses.append(error)

# Plotting the RMSE values
plt.plot(rmses)
plt.title('RMSE for the 0th to 150th harmonic over the domain [-4, 4]')
plt.xlabel('Harmonic Number')
plt.ylabel('RMSE')
plt.show()
```

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Function to generate a complex sinusoidal signal
def generate_complex_signal(frequency, duration, sampling_rate):
    # Generate time values
    t = np.arange(0, duration, 1/sampling_rate)
    # Generate complex sinusoidal signal
    signal = np.exp(2j * np.pi * frequency * t)
    return t, signal

# Function to plot the original and sampled signals in the frequency domain
def plot_frequency_domain(signal, sampling_rate):
    # Get the length of the signal
    n = len(signal)
    # Compute frequencies and FFT values
    frequencies = np.fft.fftfreq(n, d=1/sampling_rate)
    fft_values = np.fft.fft(signal)

    # Plotting the real part of the signal
    plt.figure(figsize=(10,5))
    plt.subplot(1, 2, 1)
    plt.plot(np.real(signal))
    plt.title('Real Part of Signal')

    # Plotting the frequency domain representation
    plt.subplot(1, 2, 2)
    plt.stem(frequencies, np.abs(fft_values))
    plt.title('Frequency Domain')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')

    plt.tight_layout()
    plt.show()

# Demonstrate aliasing with a complex sinusoidal signal
frequency_original = 10 
duration = 1  

# Case 1
sampling_rate_sufficient = 50  # Hz
t_sufficient, signal_sufficient = generate_complex_signal(frequency_original, duration, sampling_rate_sufficient)

# Case 2
sampling_rate_insufficient = 15  # Hz
t_insufficient, signal_insufficient = generate_complex_signal(frequency_original, duration, sampling_rate_insufficient)

# Display original signal frequency
print(f"Original Signal Frequency: {frequency_original} Hz")

# Display and plot Case 1: Sufficient Sampling Rate
print(f"\nCase 1: Sufficient Sampling Rate ({sampling_rate_sufficient} Hz)")
plot_frequency_domain(signal_sufficient, sampling_rate_sufficient)

# Display and plot Case 2: Insufficient Sampling Rate
print(f"\nCase 2: Insufficient Sampling Rate ({sampling_rate_insufficient} Hz)")
plot_frequency_domain(signal_insufficient, sampling_rate_insufficient)
```

```python
# Find the Taylor series expansion centered at x = pi/4 with 50 terms
taylor_series = sp.series(f, x, sp.pi/4, 50).removeO()

# Convert the symbolic expression to a numerical function
taylor_fn = sp.lambdify(x, taylor_series, 'numpy')

# Generate x values for the plot
x_values = np.linspace(-4, 4, 1000)

# Generate y values using the Taylor series function
y_values_taylor = taylor_fn(x_values)

# Generate y values for the original function
y_values_original = np.sin(x_values)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values_original, label=r'$\sin x$', linestyle='dashed', linewidth=2)
plt.plot(x_values, y_values_taylor, label='Taylor Series (50 terms)')
plt.title('Taylor Series of $\\sin x$ at $x = \\frac{\pi}{4}$')
plt.xlabel('x')
plt.ylabel('y')

# Adding axis lines, legend, and grid to the plot
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Display the plot
plt.show()
```

```python
# Define a 2D Gaussian filter
def gaussian_filter(size, sigma):
    x, y = np.meshgrid(np.arange(0, size, 1, dtype=float), np.arange(0, size, 1, dtype=float))
    x -= (size - 1) / 2
    y -= (size - 1) / 2
    exponent = -(x**2 + y**2) / (2 * sigma**2)
    return np.exp(exponent) / (2 * np.pi * sigma**2)

# Read the original image in grayscale
original_image = cv2.imread('4th Q Image.png', cv2.IMREAD_GRAYSCALE)

# Apply 2D Fourier transform
f_transform = np.fft.fft2(original_image)
f_transform_shifted = np.fft.fftshift(f_transform)

# Ensure the Gaussian filter size matches the size of the frequency domain image
gaussian_filter_size = min(original_image.shape[0], original_image.shape[1])
gaussian_sigma = 10
gaussian_filter = gaussian_filter(gaussian_filter_size, gaussian_sigma)

# Ensure that the Gaussian filter has the same shape as the shifted Fourier transform
gaussian_filter = np.pad(gaussian_filter, ((0, f_transform_shifted.shape[0] - gaussian_filter.shape[0]),
                                           (0, f_transform_shifted.shape[1] - gaussian_filter.shape[1])),
                         mode='constant', constant_values=0)

# Apply Gaussian filter in the frequency domain
f_transform_shifted_blurred = f_transform_shifted * gaussian_filter

# Apply inverse Fourier transform to get the blurred image
blurred_image = np.fft.ifft2(np.fft.ifftshift(f_transform_shifted_blurred)).real

# Display the original image and blurred image
plt.subplot(121), plt.imshow(original_image, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image (Gaussian Filter)'), plt.xticks([]), plt.yticks([])
 
# Show the plot
plt.show()
```

```python
# Generate x values
x = np.linspace(-5, 5, 1000)

# Define the logistic function and its derivative
def logi_function(x):
    return 1 / (1 + np.exp(-x))

def derivative_logi_function(x):
    return logi_function(x) * (1 - logi_function(x))

# Calculate the derivative of the logistic function
y = derivative_logi_function(x)

# Plotting the derivative of the logistic function
plt.figure(figsize=(10, 8))
plt.plot(x, y, label="Derivative of Logistic function")
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("f'(x) = f'(x) * (1 - f'(x))")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
